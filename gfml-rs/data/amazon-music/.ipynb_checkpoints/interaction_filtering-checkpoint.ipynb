```python
from collections import defaultdict
import pickle

with open('./train.pkl', 'rb') as f:
    user, item, rating = pickle.load(f)
    
with open('./test.pkl', 'rb') as f:
    te_user, te_item, te_rating = pickle.load(f)
    
all_item = item + te_item
all_user = user + te_user
```

```python
idx2item = {}
for i, k in enumerate(list(set(all_item))):
    idx2item[i] = k
item2idx = dict([(v, k) for k, v in idx2item.items()])
```

```python
idx2user = {}
for i, k in enumerate(list(set(all_user))):
    idx2user[i] = k
user2idx = dict([(v, k) for k, v in idx2user.items()])
```

```python
all_dict = defaultdict(list)
for u, i in zip(all_user, all_item):
    all_dict[user2idx[u]].append(item2idx[i])
```

```python
ui_dict = defaultdict(list)
for u, i in zip(user, item):
    ui_dict[user2idx[u]].append(item2idx[i])
```

```python
te_ui_dict = defaultdict(list)
for u, i in zip(te_user, te_item):
    te_ui_dict[user2idx[u]].append(item2idx[i])
```
